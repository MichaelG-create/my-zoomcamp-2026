```python
import duckdb
import kagglehub
from pathlib import Path
import time

# T√©l√©charger depuis Kaggle
print("üì• T√©l√©chargement du dataset Kaggle...")
start_time = time.time()
dataset_path = kagglehub.dataset_download(
    "bhanuthakurr/cleaned-contoso-dataset"
)
download_time = time.time() - start_time
print(f"‚úì Dataset t√©l√©charg√© en {download_time:.2f}s\n")

# Cr√©er une connexion DuckDB persistante
print("üîå Cr√©ation de la base DuckDB...")
con = duckdb.connect('accelerateur-sql.duckdb')
print("‚úì Connect√©!\n")

# Charger les fichiers CSV
print("üìÇ Chargement des donn√©es...")
csv_files = list(Path(dataset_path).glob('*.csv'))
print(f"üìÅ {len(csv_files)} fichiers CSV trouv√©s\n")

total_rows = 0
total_start = time.time()

for idx, csv_file in enumerate(csv_files, 1):
    table_name = csv_file.stem

    print(f"[{idx}/{len(csv_files)}] '{table_name}'...", end=" ", flush=True)

    # Charger directement avec DuckDB
    con.execute(f"""
        CREATE OR REPLACE TABLE {table_name} AS
        SELECT * FROM read_csv_auto('{csv_file}')
    """)

    # Compter les lignes
    row_count = con.execute(f"SELECT COUNT(*) FROM {table_name}").fetchone()[0]
    col_count = len(con.execute(f"SELECT * FROM {table_name} LIMIT 1").description)

    print(f"‚úì ({row_count:,} lignes, {col_count} colonnes)")
    total_rows += row_count

total_time = time.time() - total_start

print("\n" + "=" * 60)
print(f"‚úì COMPLET!")
print(f"  ‚Ä¢ Tables cr√©√©es: {len(csv_files)}")
print(f"  ‚Ä¢ Lignes totales: {total_rows:,}")
print(f"  ‚Ä¢ Temps total: {total_time:.2f}s")
print("=" * 60)
```

```python
%%sql
-- Premi√®re piste: segmenter par volume de commandes

-- calculer le nb de commandes par client

WITH online_sales_count_per_customer AS
  (SELECT
    CustomerKey,
    COUNT(DISTINCT OnlineSalesKey) AS online_sales_count
  FROM FactOnlineSales
  GROUP BY CustomerKey)

-- calculer le nb de clients par nb de commandes

SELECT
  online_sales_count,
  COUNT(DISTINCT CustomerKey) AS customer_count
FROM online_sales_count_per_customer
GROUP BY online_sales_count
ORDER BY online_sales_count
```

```python
%%sql
-- segmenter par panier moyen

-- calculer le panier moyen par client
WITH avg_sales_amount_per_customer AS
(SELECT
  CustomerKey,
  AVG(SalesAmount) AS avg_sales_amount
FROM FactOnlineSales
GROUP BY CustomerKey)

-- seuils
-- >= 300: gros
-- >=200= interm√©diaire
-- petit

SELECT
  CASE
    WHEN avg_sales_amount >= 300 THEN '1. High'
    WHEN avg_sales_amount >= 200 THEN '2. Medium'
    ELSE '3. Low'
  END AS customer_segment,
  COUNT(CustomerKey) AS customer_count
FROM avg_sales_amount_per_customer
GROUP BY customer_segment
ORDER BY customer_segment
```

```python
%%sql
-- le pourcentage de chaque vente par rapport au CA global

SELECT
  StoreKey,
  SalesAmount,
  SUM(SalesAmount) OVER () AS total_sales_amount,
  (SalesAmount / SUM(SalesAmount) OVER ()) * 100 AS sales_percentage
FROM FactSales
```

```python
%%sql
-- le pourcentage de chaque vente par rapport au CA du magasin

SELECT
  StoreKey,
  SalesAmount,
  SUM(SalesAmount) OVER (PARTITION BY StoreKey) AS store_sales_amount,
  (SalesAmount / SUM(SalesAmount) OVER (PARTITION BY StoreKey)) * 100 AS sales_percentage
FROM FactSales
```

```python
%%sql
-- r√©compenser les meilleurs vendeurs par magasin
-- 3 meilleures ventes par magasin
WITH sales_rank AS
(
SELECT
  StoreKey,
  SalesAmount,
  ROW_NUMBER() OVER (PARTITION BY StoreKey ORDER BY SalesAmount DESC) AS sales_rank
FROM FactSales
ORDER BY StoreKey, sales_rank
)


SELECT
  StoreKey,
  SalesAmount,
  sales_rank
FROM sales_rank
WHERE sales_rank <= 3
ORDER BY StoreKey, sales_rank
```

```python
%%sql
SELECT
  StoreKey,
  SalesAmount,
  ROW_NUMBER() OVER (PARTITION BY StoreKey ORDER BY SalesAmount DESC) AS sales_rank
FROM FactSales
QUALIFY ROW_NUMBER() OVER (PARTITION BY StoreKey ORDER BY SalesAmount DESC) <= 3
ORDER BY StoreKey, sales_rank
```
